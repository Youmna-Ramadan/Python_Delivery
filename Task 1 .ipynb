````python
#function for remove digits from data  
def remove_digits(data):
    data =''.join([i for i in data if not i.isdigit()])
    return data

#function for remove parenthesis from data  
def remove_parenthesis(data):
    x = data.find('(')
    if x>-1: data = data[:x]
    return data  
````

````python
#Replace missing values with NANs
Energy = Energy.replace('...', np.nan)

#convert from petajoule to gigajoules
Energy['Energy Supply']= Energy['Energy Supply']*1000000 

#Rename some countries
Energy = Energy.replace("Republic of Korea", "South Korea")
Energy = Energy.replace("United States of America", "United States")
Energy = Energy.replace("United Kingdom of Great Britain and Northern Ireland","United Kingdom")
Energy = Energy.replace("China, Hong Kong Special Administrative Region", "Hong Kong")

#remove numbers and parenthesis from countries names
Energy['Country']=Energy['Country'].apply(remove_digits)
Energy['Country']=Energy['Country'].apply(remove_parenthesis)
````

What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This step should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

````python
#function for calculate the average GDP 
def avgGDP():
    Top15_Countries = mergedData1[['2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']]
    Top15_Countries = Top15_Countries.dropna()
    avgGDP = Top15_Countries.mean(axis=1).sort_values(ascending= False)
    avgGDP = pd.Series(avgGDP)
    return avgGDP
````

### Step 8
Create a new column that is the ratio of Self-Citations to Total Citations. 
What is the maximum value for this new column, and what country has the highest ratio?

*This step should return a tuple with the name of the country and the ratio.*

````python
def max_Ratio():
    Top15_countries = mergedData1 
    total_citations = Top15_countries['Self-citations'].sum()
    Top15_countries['Ratio'] = Top15_countries['Self-citations'] /total_citations
    mx = Top15_countries['Ratio'].max()
    result = Top15_countries[Top15_countries['Ratio'] == mx]
    return (result.index[0], mx)
   
````

### Step 9
Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

*This step should return a series named `HighRenew` whose index is the country name sorted in ascending order of rank.*

````python
def highRenewable():
    top15_countries = mergedData1
    med = top15_countries['% Renewable'].median()
    top15_countries['HighRenew'] = (top15_countries['% Renewable'] >= med) * 1
    top15_countries.sort_values('Rank')
    top15_countries = pd.Series(top15_countries['HighRenew'])
    return top15_countries

````

Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe',

                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',

```

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
#create dictionary
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
````

````python
def countries_per_Continent():
        top15_countries = mergedData1
        top15_countries['Estimated_pop'] = top15_countries.iloc[:,7]/top15_countries.iloc[:,8]
        total_pop =top15_countries['Estimated_pop'].astype('float64')
        
        grp_Continent = total_pop.groupby(ContinentDict)
        
        size = grp_Continent.size()
        sum = grp_Continent.sum()
        mean = grp_Continent.mean()
        std = grp_Continent.std()
   
        result = pd.DataFrame({'size':size, 'sum': sum, 'mean':mean,'std':std})
        return result
````
